```python
import prop

RGP = open("RGP", "w")


fluid_in = 'CO2'

Tmin = 388.8888992
Tmax = 400.0000106
NT = 2
dT = (Tmax-Tmin)/(NT-1)

Pmin = 103421.1035156
Pmax = 137894.8046875
NP = 2
dP = (Pmax-Pmin)/(NP-1)


T=[Tmin]
for i in range(NT-1):
    T.append(T[i]+dT)

P=[Pmin]
for i in range(NP-1):
    P.append(P[i]+dP)



def ww(x1,x2):
    RGP.write("{} {}".format(x1,x2))
    RGP.write("\n")

def w(x):
    RGP.write(str(x))
    RGP.write("\n")

def wc(x):
    RGP.write("{} ".format(x))


fluid = "REFPROP::{}".format(fluid_in)
###########################################
w("$$$$HEADER")  #HEADER
w("$$${}".format(fluid_in))  #COMPONENT
w("$$PARAM")
w("27")     #количество параметров ниже
#########################################

w("DESCRIPTION")
w("Fluid by Maximov")

w("NAME")
w("$$${}".format(fluid_in))

w("INDEX")
w("$$${}".format(fluid_in))

w("MODEL")
w("3")  #superheated

w("UNITS")
w("1")  #(kg, m, s, K)

w("PMIN_SUPERHEAT")
w(Pmin)

w("PMAX_SUPERHEAT")
w(Pmax)

w("TMIN_SUPERHEAT")
w(Tmin)

w("TMAX_SUPERHEAT")
w(Tmax)

w("TMIN_SATURATION")
w(Tmin)

w("TMAX_SATURATION")
w(prop.crit(fluid)["T"])

w("SUPERCOOLING")
w(str(0))

w("P_CRITICAL")
w(prop.crit(fluid)["P"])

w("P_TRIPLE")
w(prop.triple(fluid)["P"])

w("T_CRITICAL")
w(prop.crit(fluid)["T"])

w("T_TRIPLE")
w(prop.triple(fluid)["T"])

w("GAS_CONSTANT")
w(prop.R(fluid))

w("TABLE_1")
ww(NT, NP)

w("TABLE_2")
ww(NT, NP)

w("TABLE_3")
ww(NT, NP)

w("TABLE_4")
ww(NT, NP)

w("TABLE_5")
ww(NT, NP)

w("TABLE_6")
ww(NT, NP)

w("TABLE_7")
ww(NT, NP)

w("TABLE_8")
ww(NT, NP)

w("TABLE_9")
ww(NT, NP)

w("SAT_TABLE")
w(NP) #3 значения?

w("$$SUPER_TABLE")
RGP.write("{} {} {}\n".format(NP,4,9))


w("$TABLE_1") #H
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["H"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["H"])
RGP.write("\n")

w("$TABLE_2") #A
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["A"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["A"])
RGP.write("\n")

w("$TABLE_3") #v
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["v"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["v"])
RGP.write("\n")

w("$TABLE_4") #cv
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["cv"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["cv"])
RGP.write("\n")

w("$TABLE_5") #cp
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["cp"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["cp"])
RGP.write("\n")

w("$TABLE_6") #dPdv
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["dPdv"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["dPdv"])
RGP.write("\n")

w("$TABLE_7") #s
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["s"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["s"])
RGP.write("\n")

w("$TABLE_8") #mu
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["mu"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["mu"])
RGP.write("\n")

w("$TABLE_9") #lamda
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["lamda"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["lamda"])
RGP.write("\n")
   


# w('$$SAT_TABLE')
# RGP.write("{} {} {}\n".format(NP,4,9))





RGP.close()
```
