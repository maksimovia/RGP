```python
#Только для SUPERHEATING
#Перечень из REFPROP
fluid_in = 'CO2'

#T - °C; P - МПа
Tmin = 100
Tmax = 200
NT = 100

Pmin = 0.1
Pmax = 0.3
NP = 100

Tmin = Tmin + 273.15
Tmax = Tmax + 273.15
Pmin = Pmin*10**6
Pmax = Pmax*10**6
dT = (Tmax-Tmin)/(NT-1)
dP = (Pmax-Pmin)/(NP-1)
T=[Tmin]
for i in range(NT-1):
    T.append(T[i]+dT)
P=[Pmin]
for i in range(NP-1):
    P.append(P[i]+dP)

import prop
RGP = open("{}.rgp".format(fluid_in), "w")
def ww(x1,x2):
    RGP.write("{} {}".format(x1,x2))
    RGP.write("\n")

def w(x):
    RGP.write(str(x))
    RGP.write("\n")

def wc(x):
    RGP.write("{} ".format(x))

fluid = "REFPROP::{}".format(fluid_in)
###########################################
w("$$$$HEADER")  #HEADER
w("$$${}".format(fluid_in))  #COMPONENT
w("$$PARAM")
w("24")     #количество параметров ниже

w("NAME")
w(fluid_in)

w("INDEX")
w(fluid_in)

w("MODEL")
w("3")  #superheated

w("UNITS")
w("1")  #(kg, m, s, K)

w("PMIN_SUPERHEAT")
w(Pmin)

w("PMAX_SUPERHEAT")
w(Pmax)

w("TMIN_SUPERHEAT")
w(Tmin)

w("TMAX_SUPERHEAT")
w(Tmax)

w("TMIN_SATURATION")
w(Tmin)

w("TMAX_SATURATION")
w(Tmin)

w("P_CRITICAL")
w(prop.crit(fluid)["P"])

w("P_TRIPLE")
w(prop.triple(fluid)["P"])

w("T_CRITICAL")
w(prop.crit(fluid)["T"])

w("T_TRIPLE")
w(prop.triple(fluid)["T"])

w("GAS_CONSTANT")
w(prop.R(fluid))

w("TABLE_1")
ww(NT, NP)

w("TABLE_2")
ww(NT, NP)

w("TABLE_3")
ww(NT, NP)

w("TABLE_4")
ww(NT, NP)

w("TABLE_5")
ww(NT, NP)

w("TABLE_6")
ww(NT, NP)

w("TABLE_7")
ww(NT, NP)

w("TABLE_8")
ww(NT, NP)

w("TABLE_9")
ww(NT, NP)

###########################################
w("$$$$DATA")
w("$$${}".format(fluid_in))
w("1")
w("$$PARAM")
w("24")

w("NAME")
w(fluid_in)

w("INDEX")
w(fluid_in)

w("MODEL")
w("3")

w("UNITS")
w("1")

w("PMIN_SUPERHEAT")
w(Pmin)

w("PMAX_SUPERHEAT")
w(Pmax)

w("TMIN_SUPERHEAT")
w(Tmin)

w("TMAX_SUPERHEAT")
w(Tmax)

w("TMIN_SATURATION")
w(Tmin)

w("TMAX_SATURATION")
w(Tmin)

w("P_CRITICAL")
w(prop.crit(fluid)["P"])

w("P_TRIPLE")
w(prop.triple(fluid)["P"])

w("T_CRITICAL")
w(prop.crit(fluid)["T"])

w("T_TRIPLE")
w(prop.triple(fluid)["T"])

w("GAS_CONSTANT")
w(prop.R(fluid))

w("TABLE_1")
ww(NT, NP)

w("TABLE_2")
ww(NT, NP)

w("TABLE_3")
ww(NT, NP)

w("TABLE_4")
ww(NT, NP)

w("TABLE_5")
ww(NT, NP)

w("TABLE_6")
ww(NT, NP)

w("TABLE_7")
ww(NT, NP)

w("TABLE_8")
ww(NT, NP)

w("TABLE_9")
ww(NT, NP)

w("$$SUPER_TABLE")
w(9)

w("$TABLE_1") #H
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["H"])
RGP.write("\n")
for i in range(NP):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NP):
    wc(prop.p_q(P[i],1,fluid)["H"])
RGP.write("\n")
print("1+")

w("$TABLE_2") #A
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["A"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["A"])
RGP.write("\n")
print("2+")

w("$TABLE_3") #v
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["v"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["v"])
RGP.write("\n")
print("3+")

w("$TABLE_4") #cv
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["cv"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["cv"])
RGP.write("\n")
print("4+")

w("$TABLE_5") #cp
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["cp"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["cp"])
RGP.write("\n")
print("5+")

w("$TABLE_6") #dPdv
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["dPdv"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["dPdv"])
RGP.write("\n")
print("6+")

w("$TABLE_7") #s
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["s"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["s"])
RGP.write("\n")
print("7+")

w("$TABLE_8") #mu
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["mu"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["mu"])
RGP.write("\n")
print("8+")

w("$TABLE_9") #lamda
ww(NT,NP)
for i in range(NT):
    wc(T[i])
RGP.write("\n")
for i in range(NP):
    wc(P[i])
RGP.write("\n")
for j in range(NP):
    for i in range(NT):
        wc(prop.t_p(T[i],P[j],fluid)["lamda"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["T"])
RGP.write("\n")
for i in range(NT):
    wc(prop.p_q(P[i],1,fluid)["lamda"])
RGP.write("\n")
print("9+")

RGP.close()
print("++")
```
